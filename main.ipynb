```python
data.drop(columns=['MiscFeature','Fence','PoolQC','Alley','FireplaceQu','MasVnrType'], inplace=True)
```

```python
cat_col = data.select_dtypes(include = 'object').columns
```

```python
data[cat_col].isna().sum()
```

```python
data
```

```python
cat_imputer = SimpleImputer(strategy= 'most_frequent')
```

```python
data[cat_col] = cat_imputer.fit_transform(data[cat_col])
```

```python
data[cat_col].isna().sum()
```

```python
data.isna().sum()
```

```python
len(numerical_col)
len(cat_col)
```

```python
numerical_col = numerical_col.drop('SalePrice')
```

```python
target_col = 'SalePrice'
```

```python
from sklearn.feature_selection import SelectKBest , r_regression , f_classif
```

```python
k_best = SelectKBest(score_func=r_regression , k = 30)
k_best_num = k_best.fit_transform(data[numerical_col] , data[target_col])
```

```python
temp = data[numerical_col]
```

```python
k_best_num_fea = temp.columns[k_best.get_support()]
```

```python
k_best_num_fea
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder(handle_unknown='ignore')

encoder.fit(data[cat_col])
encoded_Data = encoder.transform(data[cat_col])
encoded_Data.shape
encoded_dataframe = pd.DataFrame.sparse.from_spmatrix(encoded_Data , columns=encoder.get_feature_names_out())
encoded_dataframe
```

```python
input_datframe = pd.concat([data[k_best_num_fea] , encoded_dataframe] , axis=1)
input_datframe
```

```python
output_dataframe = data[target_col]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression(n_jobs= -1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(input_datframe , output_dataframe , test_size=0.2 , random_state=42)
```

```python
from sklearn.model_selection import RepeatedKFold , cross_val_score
```

```python
kfold = RepeatedKFold(n_splits=10 , n_repeats=3 , random_state=42 )
socre_val = cross_val_score(model , x_train , y_train , cv = kfold , scoring = 'r2' , n_jobs = -1 )
```

```python
print('Accuracy: %.3f (%.3f)' % (np.mean(socre_val), np.std(socre_val)))
```

```python
socre_val
```

```python
model.fit(x_train , y_train)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
y_pred = model.predict(x_test)

# Calculate metrics
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)
print("R-squared:", r2)
```

```python
from sklearn.linear_model import Ridge

ridge = Ridge(alpha=1.0)
ridge.fit(x_train , y_train)
```
